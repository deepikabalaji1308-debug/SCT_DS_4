```python
#Step 1:Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
#Step 2:Load Dataset
df=pd.read_csv("/content/traffic_accident_dataset.csv")
df.head()
```

```python
#Step 3:Dataset Information
print(df.shape)
df.info()
df.describe()
```

```python
#Step 4:Checking Missing Values
df.isnull().sum()
```

```python
#Step 5:Weather Conditions
plt.figure(figsize=(10,6))
sns.countplot(y=df['Weather_Condition'],order=df['Weather_Condition'].value_counts().head(10).index)
plt.title("Top 10 Weather Conditions")
plt.show()
```

```python
#Step 6:Read Conditions
plt.figure(figsize=(8,5))
sns.countplot(x='Road_Condition',data=df)
plt.title("Road Conditions")
plt.show()
```

```python
#Step 7:Accident Severity
sns.countplot(x='Severity',data=df)
plt.title("Accident Severity")
plt.show()
```

```python
#Step 8:Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Step 9:Accident Hotspots
plt.figure(figsize=(8,6))
plt.scatter(df['Start_Lng'],df['Start_Lat'],alpha=0.2)
plt.title("Accident Hotspots")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
